```python
## 2.1	Analysis of Literacy rate vs Total crimes

plt.figure(figsize=(10,8))
plt.scatter(df["literacy_rate"], df["Total_crime"], color="red", marker='+')
plt.title("literacy_rate & Total_crime by statewise",fontsize = 20)
plt.xlabel("literacy_rate")
plt.ylabel("Total_crime")
plt.show()
```

```python
##2.4	 Analysis of area vs overall crime

plt.figure(figsize=(10,8))
plt.scatter(df["Area"], df["Total_crime"], color="red", marker='+')
plt.title("Area vs Total_crime by statewise",fontsize = 20)
plt.xlabel("Area")
plt.ylabel("Total_crime")
plt.show()
```

```python
##2.5	 Analysis of population vs overall crime

plt.figure(figsize=(10,8))
plt.scatter(df["Population"], df["Total_crime"], color="red", marker='+')
plt.title("Population & Total_crime by statewise",fontsize = 20)
plt.xlabel("Total_crime")
plt.ylabel("Population")
plt.show()
```
